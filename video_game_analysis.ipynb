## Performing thorough data cleaning of the dataset

```python
df = pd.read_csv('Global_Video_Game_Sales.csv')

df.isnull().sum()              # count missing values per column

df.drop_duplicates()           # remove duplicate rows

df.dropna()                    # remove missing values

df.reset_index(drop=True)      # reset index after cleaning
```

 ## Variables of interest to work with

```python
variables = data[["Year_of_Release", "NA_Sales", "EU_Sales", "JP_Sales", "Other_Sales", "Global_Sales", "Critic_Score"]]
```

```python
variables
```

## Plot boxplots to identify whether we have outliers

```python
correlation = variables.corr()    # calculate correlation between columns
plt.figure(figsize=(10,8))        # set figure size
sns.heatmap(correlation, annot=True,cmap="coolwarm", linewidth = 0.5)    # create heatmap with values
plt.title("Global Video Game Sales & Ratings Correlation HeatMap")       # add title
plt.show()  # display the plot
```

```python
plt.figure(figsize=(10,8))                              # set figure size
sns.boxplot(x =variables["Year_of_Release"],            # create boxplot
            y=variables["Critic_Score"])                 
plt.xlabel("Year_of_Release")                           # label x-axis
plt.ylabel("Critic_Score")
plt.title("Boxplot of Year_of_Release by Critic_Score") # add title
plt.show()                                              # display the plot
```

## Removing the outliers and plot new boxplots to see the results

```python
cleaned_data = variables[(variables ["Critic_Score"] >=70)& (variables ["Critic_Score"] <=95)]  # filter data to keep Critic_Score between 70 and 95
plt.figure(figsize = (10,8))   # set figure size
sns.boxplot(x = cleaned_data["Year_of_Release"], y = cleaned_data["Critic_Score"]) # create boxplot without outliers
plt.xlabel("Year_of_Release")  # label x-axis
plt.ylabel("Critic_Score")     # label y-axis
plt.title("Data Amputation to remove outliers")  # add title
plt.show() # display the plot
```

## Results seen in the boxplots.

#### Boxplots before Cleaning:

#### The first charts looked really squished.The main box (which shows where most games fall) was tiny and stuck to the bottom line. This tells us that most video games only sell a small amount. The most important thing was the many tiny dots scattered far above the main box. These dots are called outliers. They are the super-hit games that sold much, much more than every other game. They make the chart hard to read.

#### Boxplots after data amputation:

#### I fixed the data by using a process called data amputation (which just means cutting out the extreme outliers). The new charts look much cleaner and easier to read. The extreme dots are now gone because the huge-selling games were removed, the vertical scale of the chart changed. We can now see the details of the normal games sales much more clearly.

## Description of the steps used in the project.

#### I followed seven simple steps to finish the assignment:

#### 1. Load the Data: I put the game sales file into the Python Notebook.

#### 2. Check the Data: I checked the size of the table (how many rows and columns) and got the average scores and sales numbers.

#### 3. Clean the Data: I fixed the data in two ways, I remove duplicate rows, missing information and I reset index after cleaning.

#### 4. Pick Variables: I chose 7 columns, focusing on the sales numbers and the critic/user scores.

#### 5. Plot Boxplots: I made the first charts to find the outliers (the extreme high sales games).

#### 6. Remove Outliers: I found the normal sales and any game outside these values was removed from the data. This removal is the data amputation step.

#### 7. Plot Boxplots: I made the final charts to show that the outliers were successfully deleted, making the data much cleaner.
